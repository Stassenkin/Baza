```python
documents = [
        {"type": "passport", "number": "2207 876234", "name": "Василий Гупкин"},
        {"type": "invoice", "number": "11-2", "name": "Геннадий Покемонов"},
        {"type": "insurance", "number": "10006", "name": "Аристарх Павлов"},
      ]
 
directories = {
        '1': ['2207 876234', '11-2'],
        '2': ['10006'],
        '3': []
      }
 
def find_people(documents_list):
    '''
    Help to find men from his documents
    '''
    number_of_doc = str(input('Введите пожалуйста номер документа '))
    index = 1
    x= len(documents_list)
    for doc in documents_list:
        if number_of_doc == doc["number"]:
            print(doc["name"])
            break
        elif index == x and number_of_doc != doc["number"]:
            print ('вы ввели несуществующий документ')
        index +=1   
 
#people(documents)
 
def print_all_list(documents_list):
    for doc in documents_list:
        print(doc['type'],doc['number'],doc['name'])
#list(documents)
 
def take_from_shelf(direct,documents_list):
    number_of_doc = str(input('Введите пожалуйста номер документа '))
    for shelf in direct:
        for doc_on_the_shelf in direct[shelf]:
            if doc_on_the_shelf == number_of_doc:
                print('Документ на полке № ', shelf)
    index = 1
    x= len(documents_list)            
    for doc in documents_list:
        if number_of_doc == doc["number"]:
            break
        elif index == x and number_of_doc != doc["number"]:
            print ('результат не найден в базе')
        index +=1   
 
 
#shelf(directories)
 
def add_to_list(documents_list,direct):
    type_1 = str(input('введите тип документа '))
    number_1 = str(input('введите номер документа '))
    name_1 = str(input('введите имя владельца '))
    shelf_1 = str(input('введите номер полки на которую поставить '))
    dict_1= dict(type = type_1, number = number_1, name = name_1)
    documents.append(dict_1)
    index=1
    x= len(direct)
    for shelf in direct:
        if shelf_1 == shelf:
            direct[shelf].append(number_1)
            break
        else:
            if index == x:
                direct[shelf_1]=[number_1]
                break    
        index +=1
    print(direct)
    

 
#add(documents,directories)
 
def print_all_name(documents_list):
    for doc in documents_list:
        try:
            if "number" in doc:
                doc['name']
                print (doc['name'])
        except KeyError:
            print(' у документа нет владельца Ошибка KeyError  ')
def add_shelf():
    user_input = input ('Введите номер полки:')
    if user_input in directories:
        print ('Такая полка уже существует. Текущий перечень полок', ','.join(directories.keys()))
        return
    else: 
        directories[user_input] = []
        print ('Полка добавлена. Текущий перечень полок:', ',' .join(directories.keys()))
        return
def delete_shelf():
    user_input = input('Введите номер полки ')
    if user_input not in directories.keys():
        print ('Такой полки не существует. Текущий перечень полок: ', *directories.keys())
        return
    else:
        del(directories[user_input])
        print ('Полка удалена. Текущий перечень полок: ', *directories.keys())
        return
 
#list(documents)
 
def main():
    while True:
        command=input('\nВведите команду\n'
                "p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;\n"
                'l– list – команда, которая выведет список всех документов в формате passport "2207 876234" "Василий Гупкин";\n'
                "s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится;\n"
                "as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень;\n"
                'ds - delete shelf – команда, которая спросит номер документа и удалит его из каталога и из перечня полок;\n'
                'q - Выйти из программы\n'
                      )
        if command =='p':
            find_people(documents)
        elif command =='l':
            print_all_list(documents)
        elif command =='s':
            take_from_shelf(directories,documents)
        elif command =='a':
            add_to_list(documents,directories)
        elif command =='n':
            print_all_name(documents)
        elif command =='as':
            add_shelf()
        elif command =='ds':
            delete_shelf()
        elif command =='q':
            print('До свидания')
main()
```
